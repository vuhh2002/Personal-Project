```python
def Max():
    def forward(x):
        res = max(x)
        return res
    
    return Benchmark(
        forward = forward,
        name = 'Max',
    )
Max = Max()
Max.show(-100, 100)
```

```python
def ProdLogSigmoid():
    def forward(x):
        x = th.tensor(x)
        log_sigmoid = np.log(1 + th.sigmoid(x))
        res = log_sigmoid.prod()
        return res
    
    return Benchmark(
        forward = forward,
        name = 'ProdLogSigmoid',
    )
ProdLogSigmoid = ProdLogSigmoid()
ProdLogSigmoid.show(-10, 10)
```

```python
def SumLogSigmoid():
    def forward(x):
        x = th.tensor(x)
        log_sigmoid = np.log(th.sigmoid(x))
        res = sum(log_sigmoid)
        return res
    
    return Benchmark(
        forward = forward,
        name = 'SumLogSigmoid',
    )
SumLogSigmoid = SumLogSigmoid()
SumLogSigmoid.show(-10, 10)
```

```python
def SumSigmoid():
    def forward(x):
        x = th.tensor(x)
        sigmoid = th.sigmoid(x)
        res = sum(sigmoid)
        return res
    
    return Benchmark(
        forward = forward,
        name = 'SumSigmoid',
    )
SumSigmoid = SumSigmoid()
SumSigmoid.show(-5, 5)
```

```python
def LogSoftmax():
    def forward(x):
        x = th.tensor(x)
        log_softmax = th.log_softmax(x, dim=0)
        res = sum(log_softmax)
        return res
    
    return Benchmark(
        forward = forward,
        name = 'LogSoftMax',
    )
LogSoftmax = LogSoftmax()
LogSoftmax.show(-10, 10)
```

```python
def Sum():
    def forward(x):
        return np.sum(x)
    
    return Benchmark(
        forward = forward,
        name = 'Sum',
    )
Sum = Sum()
Sum.show(-10, 10)
```

```python
def Product():
    def forward(x):
        return np.prod(x)
    
    return Benchmark(
        forward = forward,
        name = 'Product',
    )
Product = Product()
Product.show(0.1, 10)
```
